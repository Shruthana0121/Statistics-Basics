# **Statistics Basics**

**1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

-->
* Descriptive Statistics: These are used to summarize, organize, and describe
the basic features of a dataset. It focuses on the data at hand without making generalizations beyond it.

Example: Calculating the average test score of 30 students in a specific class.


* Inferential Statistics: These use a random sample of data to make inferences, predictions, or generalizations about a larger population.

Example: Surveying 1,000 citizens to predict the outcome of a national election.

**2. What is sampling in statistics? Explain the differences between random and stratified sampling.**

-->
* Sampling: This is the process of selecting a subset of individuals from a population to estimate characteristics of the whole group.


* Random Sampling: Every member of the entire population has an equal chance of being selected. It is unbiased but may occasionally miss small subgroups.


* Stratified Sampling: The population is divided into subgroups (strata) based on shared traits (e.g., gender, age). A random sample is then taken from each stratum to ensure all groups are represented.




**3. Define mean, median, and mode. Explain why these measures of central tendency are important.**

-->
* Mean: The arithmetic average of a dataset (sum divided by count).
* Median: The middle value when data is arranged in numerical order.
* Mode: The value that appears most frequently in a dataset.

Importance: These measures are crucial because they identify the "center" of a dataset, helping analysts understand where most data points are clustered.

**4. Explain skewness and kurtosis. What does a positive skew imply about the data?**

-->
* Skewness: Measures the lack of symmetry in a distribution.

* Kurtosis: Measures the "tailedness" or the peak of a distribution compared to a normal distribution.

* Positive Skew: Implies that the tail on the right side of the distribution is longer. This indicates that the mean is typically greater than the median, and most data points are concentrated on the lower end of the scale.

**5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.  
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**

```python
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28] # [cite: 9]

mean_val = statistics.mean(numbers)
median_val = statistics.median(numbers)
mode_val = statistics.multimode(numbers)

print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
```

**6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:  
list_x = [10, 20, 30, 40, 50]  
list_y = [15, 25, 35, 45, 60]**

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0][1]
correlation = np.corrcoef(list_x, list_y)[0][1]

print(f"Covariance: {covariance}")
print(f"Correlation: {correlation}")
```

**7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:  
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]**

```python
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] #

plt.boxplot(data)
plt.title("Boxplot for Data Analysis")
plt.ylabel("Values")
plt.show()
```

Explanation: The boxplot displays the distribution of the data through quartiles. The value 35 is an outlier as it lies significantly above the upper whisker of the plot

**8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.  
-Explain how you would use covariance and correlation to explore this relationship.  
-Write Python code to compute the correlation between the two lists:  
advertising_spend = [200, 250, 300, 400, 500]  
daily_sales = [2200, 2450, 2750, 3200, 4000]**

-->
Analysis Approach: I would use Covariance to determine the direction of the relationship (positive means both increase together). I would use Correlation to measure the strength of this relationship on a scale of -1 to 1.

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500] # [cite: 15]
daily_sales = [2200, 2450, 2750, 3200, 4000] # [cite: 15]

correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]
print(f"Correlation between Spend and Sales: {correlation}")
```

**9. Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.  
-Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) youâ€™d use.  
-Write Python code to create a histogram using Matplotlib for the survey data:  
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**

-->
* Summary Statistics: I would use the Mean to find the average satisfaction score and Standard Deviation to understand how much the scores vary from that average.

* Visualization: A Histogram would be used to see the frequency of specific scores and identify the shape of the data (e.g., bell-shaped or skewed).

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7] # [cite: 17]

plt.hist(survey_scores, bins=range(1, 12), color='skyblue', edgecolor='black', align='left')
plt.title("Customer Satisfaction Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```
